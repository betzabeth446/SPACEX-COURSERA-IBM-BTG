🚀 Cada lanzamiento tiene como objetivo una órbita específica
Cada lanzamiento está destinado a una órbita específica, y aquí algunos tipos de órbitas comunes:

Aquí te dejo una lista explicada de los tipos de órbitas que probablemente veas en el dataset (Orbit):

Tipo de Órbita	Significado
-LEO	Low Earth Orbit: Órbita baja terrestre (hasta 2.000 km). Usada para satélites de observación, comunicación, y la EEI.
-GTO	Geostationary Transfer Orbit: Órbita de transferencia hacia una órbita geoestacionaria.
-GEO	Geostationary Orbit: Órbita geoestacionaria. Satélites permanecen sobre el mismo punto del ecuador.
-PO	Polar Orbit: Satélite cruza por los polos de la Tierra, usado para monitoreo global.
-ISS	International Space Station Orbit: Órbita específica de la EEI.
-HEO	Highly Elliptical Orbit: Órbita muy elíptica, utilizada para ciertas misiones científicas.
-MEO	Medium Earth Orbit: Usada por sistemas de navegación como GPS.
-SO	Sun-Synchronous Orbit: Mantiene la misma posición relativa al Sol, ideal para observación terrestre.

# **TAREA 3: Calcular la cantidad y ocurrencia de los resultados de misión (Outcome)**
Usa el método .value_counts() en la columna Outcome para calcular cuántas veces aparece cada resultado de aterrizaje. Luego, asigna ese resultado a la variable landing_outcomes.

```python
# landing_outcomes = values on Outcome column
landing_outcomes = spacex_df['Outcome'].value_counts()
landing_outcomes
```

True Ocean significa que el resultado de la misión fue un aterrizaje exitoso en una zona específica del océano.

False Ocean indica que el aterrizaje en una zona específica del océano fue fallido.

True RTLS significa que la misión aterrizó exitosamente en una plataforma terrestre (RTLS = Return To Launch Site).

False RTLS indica que no se logró el aterrizaje en la plataforma terrestre.

True ASDS significa que la misión aterrizó exitosamente en una nave no tripulada en el océano (Autonomous Spaceport Drone Ship).

False ASDS indica que falló el aterrizaje en la nave no tripulada.

None ASDS y None None representan un fallo total en el aterrizaje, es decir, no se logró ningún tipo de recuperación del propulsor.

```python
for i,outcome in enumerate(landing_outcomes.keys()):
    print(i,outcome)
```


Creamos un conjunto de resultados donde la segunda etapa no aterrizó exitosamente.

```python
bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

TAREA 4: Crear una etiqueta de resultado de aterrizaje desde la columna Outcome
Usando la columna Outcome, crea una lista donde el elemento sea 0 si la fila correspondiente está dentro del conjunto bad_outcomes; de lo contrario, que sea 1. Luego, asigna esa lista a la variable landing_class.



```python
## Add landing_class as a new column
spacex_df['Class'] = landing_class

# Verificar que se haya agregado correctamente
spacex_df[['Outcome', 'Class']].head()

```

Esta variable representará la variable de clasificación que indica el resultado de cada lanzamiento.
Si el valor es cero (0), significa que la primera etapa no aterrizó con éxito;
