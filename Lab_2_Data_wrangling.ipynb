# **TAREA 3: Calcular la cantidad y ocurrencia de los resultados de misi√≥n (Outcome)**

```python
# landing_outcomes = values on Outcome column
landing_outcomes = spacex_df['Outcome'].value_counts()
landing_outcomes
```

True Ocean significa que el resultado de la misi√≥n fue un aterrizaje exitoso en una zona espec√≠fica del oc√©ano.

True RTLS significa que la misi√≥n aterriz√≥ exitosamente en una plataforma terrestre (RTLS = Return To Launch Site).

False RTLS indica que no se logr√≥ el aterrizaje en la plataforma terrestre.

True ASDS significa que la misi√≥n aterriz√≥ exitosamente en una nave no tripulada en el oc√©ano (Autonomous Spaceport Drone Ship).

False ASDS indica que fall√≥ el aterrizaje en la nave no tripulada.

None ASDS y None None representan un fallo total en el aterrizaje, es decir, no se logr√≥ ning√∫n tipo de recuperaci√≥n del propulsor.
